Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++

7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import matplotlib.pyplot as plt
import numpy as np
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
positions = np.arange(len(languages))
language_with_pos = [f"{lang} \n({pos})" for lang,pos in zip(languages,positions)]

plt.bar(positions,popularity,color='skyblue',edgecolor='black',linewidth=1.5)
plt.xticks(positions,language_with_pos)
plt.xlabel("Programming Languages")
plt.ylabel('Popularity (%)')
plt.title('Popularity of programming language')
plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29) 

```python
import matplotlib.pyplot as plt
import numpy as np
men_means = [22,30,35,35,26]
women_means = [25,32,30,35,29]
groups = ['Group 1','Group 2','Group 3','Group 4','Group 5']
positions = np.arange(len(groups))
bar_width = 0.3
plt.bar(positions - bar_width / 2,men_means,width=bar_width,color='blue',label='Men',edgecolor='black',linewidth=1.5)
plt.bar(positions + bar_width / 2 , women_means, width=bar_width, color='red', label='Women', edgecolor='black', linewidth=1.5)
plt.xticks(positions,groups)
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend()
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
plt.bar(labels,mean_velocity,yerr = std_dev_velocity,capsize=5,color = 'skyblue',edgecolor='black')
plt.xlabel("Samples")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars")
plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
bars = plt.bar(labels, mean_velocity, yerr=std_dev_velocity, capsize=5, color='skyblue', edgecolor='black')
for bar, mean in zip(bars, mean_velocity):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(int(mean)), 
             ha='center', va='bottom', fontsize=12, fontweight='bold')
plt.xlabel("Samples")
plt.ylabel("Mean Velocity")
plt.title("Bar Plot with Error Bars and Labels")
plt.show()
```

Matplotlib Scatterplot

1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,color='blue',alpha=0.5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot of random distribution")
plt.show()
```

Seaborn Bar plot

1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
data = pd.read_csv("titanic.csv")
sns.countplot(x='sex',hue='survived',data=data,palette='pastel')
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Bar plot of sex vs survived")
plt.legend(title='Survived',labels = ['No','Yes'])
plt.show()
```

```python
sns.pointplot(x='sex',y='survived',data=data,capsize=0.1,color='blue')
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Point plot of sex vs survived")
plt.show()
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
import plotly.express as px
df = pd.read_csv("state.csv")
fig = px.scatter(
    df,
    x='Rank',
    y='Population',
    hover_name="State",
    hover_data=['Postal'],
    title="Population vs Rank of US State",
    labels = {"Population": "Population", "Rank": "Rank"},
)
fig.show()
```
